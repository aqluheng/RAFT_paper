```python
import torch
```

```python
sintelDict = torch.load("dataLog/sintelDataset_final.pth")
chairsDict = torch.load("dataLog/chairsDataset_final.pth")
moviDict = torch.load("dataLog/moviDataset_final.pth")
moviFilterDict = torch.load("dataLog/MoviFilterDataset_final.pth")
# thingsDict = torch.load("dataLog/thingsDataset_final.pth")
```

```python
import matplotlib.pyplot as plt
plt.plot(torch.log10(sintelDict["velocityHist"]/sintelDict["velocityHist"].sum()),label="sintel")
plt.plot(torch.log10(moviDict["velocityHist"]/moviDict["velocityHist"].sum()),label="movi")
plt.plot(torch.log10(chairsDict["velocityHist"]/chairsDict["velocityHist"].sum()),label="chairs")
plt.plot(torch.log10(moviFilterDict["velocityHist"]/moviFilterDict["velocityHist"].sum()),label="moviFilter")
plt.legend()
plt.xlabel("Displacement (0.1px)")
plt.ylabel("ratio of pixels(log scale)")
```

```python
import matplotlib.pyplot as plt
plt.plot(torch.log(sintelDict["directionHist"]/sintelDict["directionHist"].sum()),label="sintel")
plt.plot(torch.log(moviDict["directionHist"]/moviDict["directionHist"].sum()),label="movi")
plt.plot(torch.log(chairsDict["directionHist"]/chairsDict["directionHist"].sum()),label="chairs")
plt.plot(torch.log(moviFilterDict["directionHist"]/moviFilterDict["directionHist"].sum()),label="moviFilter")
plt.legend()
plt.xlabel("Angle (0.1degree)")
plt.ylabel("ratio of pixels(log scale)")
```
